```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0, 2*np.pi, 360)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y,'r-', marker='o')
```

```python
seg = np.pi/180.0
```

```python
x1 = np.linspace(2*np.pi+seg, 3*np.pi, 179)
y1 = np.zeros(x1.shape[0])
x2 = np.linspace(3*np.pi+seg, 4*np.pi, 179)
y2 = .5*x2
x3 = np.linspace(4*np.pi+seg, 5*np.pi, 179)
y3 = np.array([1]*179)
x4 = np.linspace(5*np.pi+seg, 6*np.pi, 179)
y4 = x4 +1-5*np.pi
print (x4[-1]/np.pi)
x = np.concatenate([x, x1, x2, x3, x4])
y = np.concatenate([y, y1, y2, y3, y4])
```

```python
x *= (100*np.pi)
y *= (100*np.pi)
fig, ax = plt.subplots(figsize=(22,10))
plt.plot(x,y,'r-', marker='o',ms=5)
```

```python
arr = np.array([1/48, 1/36,1/24,1/12,1/6])
```

```python
arr/= np.pi
arr *= (100*np.pi)
```

```python
arr = [2, 3, 4, 8, 9, 12, 16]
res = []
fig, ax = plt.subplots(figsize=(22,10))
for i in arr:
    res.append(pd.rolling_std(y, window=i))
    ax.plot(x, res[-1])
ax.axhline(np.std(y))
```

```python
arr = [30,50,80,100,150,200]
res = []
fig, ax = plt.subplots(figsize=(22,10))
for i in arr:
    res.append(pd.rolling_std(y, window=i))
    ax.plot(x, res[-1], label=str(i))
ax.axhline(np.std(y))
plt.legend()
```
